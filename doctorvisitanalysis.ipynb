5. Data Cleaning : Data Cleaning is an important process which removes unwanted data such as
* Missing values
* Duplicates and unwanted data

```python
df.isnull().sum()     #find out the null values in the data
```

```python
df.duplicated().sum()
```

```python
#we remove the columns 'PatientId' and 'AppointmentID' as we have no useful insights from it
df.drop(['PatientId','AppointmentID'], axis=1, inplace = True)
```

```python
#From the No-show column lets convert'Yes' and 'No' to 0 and 1 (where 0=noshow , 1=show)
df['No-show'] = df['No-show'].apply( lambda x: 1 if x == 'No' else 0)
```

```python
#convert dtypes od ScheduledDay and AppointmentDay into datetime dtypes
df['ScheduledDay']=pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay']=pd.to_datetime(df['AppointmentDay'])
```

```python
def ratio_calculate(df, column_name1, column_name2):
  ratio = df.groupby(column_name1)[column_name2].value_counts(normalize=True).unstack(column_name2)
  return ratio
```

Question : Patients who have scholarship show up or not

```python
scholarship_ratio = ratio_calculate(df,"Scholarship","No-show")
scholarship_ratio
```

```python
scholarship_ratio.plot(kind = "bar",figsize=(7,4))
plt.title('Percentage of showup Patients with scholarship Vs without scholarship', fontsize=9,color="black",weight='bold')
plt.ylabel('Atendance precentage', fontsize=9,color="black");
plt.xlabel('scholarship receivers status', fontsize=9,color="black");
locations = ['Without Scholarship','With Scolarship']
plt.xticks(np.arange(len(locations)),locations);
plt.legend(["didn't show"," showed"], fontsize=7);
```
